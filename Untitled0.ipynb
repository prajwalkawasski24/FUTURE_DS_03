```python
import pandas as pd

df = pd.read_csv('//content/student_feedback.csv')
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
rating_cols = df.columns[1:]  # Exclude Student ID
df[rating_cols] = df[rating_cols].apply(pd.to_numeric, errors='coerce')
df.dropna(inplace=True)
```

```python
df[rating_cols].describe()
```

```python
df[rating_cols].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
avg_ratings = df[rating_cols].mean().sort_values()

# Convert to DataFrame for seaborn compatibility with hue
avg_df = avg_ratings.reset_index()
avg_df.columns = ['Aspect', 'Average Rating']

sns.barplot(
    data=avg_df,
    x='Average Rating',
    y='Aspect',
    hue='Aspect',  # required for palette
    palette='coolwarm',
    dodge=False,
    legend=False  # disable legend explicitly
)

plt.title("Average Ratings per Feedback Question")
plt.xlabel("Average Rating (1 to 5)")
plt.ylabel("Feedback Aspect")
plt.grid(True, axis='x')
plt.show()
```

```python
corr_matrix = df[rating_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title("Correlation Matrix of Feedback Ratings")
plt.show()
```

```python
highs = avg_ratings[avg_ratings > 4.0]
lows = avg_ratings[avg_ratings < 3.0]

print("Strong Areas:\n", highs)
print("\nAreas to Improve:\n", lows)
```
